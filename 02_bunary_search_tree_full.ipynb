```python
class BSTNode:

    def __init__(self, key, val, parent):
        self.NodeKey = key # ключ узла
        self.NodeValue = val # значение в узле
        self.Parent = parent # родитель или None для корня
        self.LeftChild = None # левый потомок
        self.RightChild = None # правый потомок


class BSTFind: # промежуточный результат поиска

    def __init__(self):
        self.Node = None # None если не найден узел
        self.NodeHasKey = False # True если узел найден
        self.ToLeft = False # True, если родительскому узлу надо 
        # добавить новый узел левым потомком

class BST:

    def __init__(self, node):
        self.Root = node # корень дерева, или None

    def FindNodeByKey(self, key):
        #if self.Root is None:
        #    return None
        
        node = self.Root
        result = BSTFind()
        
        while node is not None:
            if node.NodeKey > key:
                node_parent = node
                node = node.LeftChild
            elif node.NodeKey < key:
                node_parent = node
                node = node.RightChild
            elif node.NodeKey == key:
                result.Node = node
                result.NodeHasKey = True
                return result
        
        if self.Root.LeftChild is None and self.Root.RightChild is None:
            result.Node = None
        else:
            result.Node = node_parent
        #result.NodeHasKey = False
        if node_parent.NodeKey > key:
            result.ToLeft = True
        return result
    
    
    def AddKeyValue(self, key, val):
        new_node = BSTNode(key, val, None)
        result = self.FindNodeByKey(key)
        if result.NodeHasKey:
            return False
        
        if result.Node is None:
            if result.ToLeft:
                self.Root.LeftChild = new_node
            else:
                self.Root.RightChild = new_node
            new_node.Parent = self.Root
            return True
                    
        if result.ToLeft:
            result.Node.LeftChild = new_node
        else:
            result.Node.RightChild = new_node
        new_node.Parent = result.Node
        return True
  
    def FinMinMax(self, FromNode, FindMax):      # тест на одном корне, тест не с корня           
        node = FromNode
        if FindMax:
            while node.RightChild is not None:
                node = node.RightChild
            return node                                 
        
        elif FindMax == False:
            while node.LeftChild is not None:
                node = node.LeftChild
            return node
        
        
         
    def DeleteNodeByKey (self, key):
        node = self.FindNodeByKey (key)
        if node is None or node.NodeHasKey == False:         #условия False None или ненайденный узел
            return False
        node = node.Node
                
        if node.LeftChild is not None and node.RightChild is not None:    #если оба потомка не None
            node_hair = node.RightChild
            while node_hair.LeftChild is not None:
                node_hair = node_hair.LeftChild
                    
            node_hair.LeftChild = node.LeftChild
            node_hair.LeftChild.Parent = node_hair
            if node_hair.Parent is not node:
                if node_hair.RightChild is not None:
                    node_hair.RightChild.Parent = node_hair.Parent
                    node_hair.Parent.LeftChild = node_hair.RightChild
                else:
                    node_hair.Parent.LeftChild = None
                node_hair.RightChild = node.RightChild
                node_hair.RightChild.Parent = node_hair
                
            node_hair.Parent = node.Parent            
            if node is self.Root:                                              #если удаляем корень
                self.Root = node_hair
            else:                                                              #если удаляем не корень
                if node.Parent.LeftChild == node:                                   #если вставляем левым потомком
                    node_hair.Parent.LeftChild = node_hair
                elif node_hair.Parent.RightChild == node:                           #если вставляем правым потомком
                    node_hair.Parent.RightChild = node_hair                        
            return True
                
        if node.LeftChild is not None:                                    #если только левый потомок не None
            if node is self.Root:                                              #если удаляем корень
                self.Root = node.LeftChild
                self.Root.Parent = None
            else:                                                              #если удаляем не корень
                node.LeftChild.Parent = node.Parent      
                if node.Parent.LeftChild == node:                                   #если вставляем левым потомком
                    node.Parent.LeftChild = node.LeftChild
                elif node.Parent.RightChild == node:                           #если вставляем правым потомком
                    node.Parent.RightChild = node.LeftChild
                
        elif node.RightChild is not None:                                 #если только правый потомок не None
            if node is self.Root:                                              #если удаляем корень
                self.Root = node.RightChild
                self.Root.Parent = None
            else:                                                              #если удаляем не корень
                node.RightChild.Parent = node.Parent 
                if node.Parent.LeftChild == node:                                   #если вставляем левым потомком
                    node.Parent.LeftChild = node.RightChild
                elif node.Parent.RightChild == node:                           #если вставляем правым потомком
                    node.Parent.RightChild = node.RightChild                
                
        else:                                                            # оба потомка None
            if node is self.Root:                                              #если удаляем корень 
                self.Root = None
            else:                                                              #если удаляем не корень
                if node.Parent.LeftChild == node:                                   #если удаляем левый потомлк
                    node.Parent.LeftChild = None
                elif node.Parent.RightChild == node:                           #если удаляем правый потомок
                    node.Parent.RightChild = None            
        return True

    def Count(self):
        if self.Root is None:
            return 0
        list_nodes = []
        list_nodes.append(self.Root)
        for i in list_nodes:
            if i.LeftChild is not None:
                list_nodes.append(i.LeftChild)
            if i.RightChild is not None:
                list_nodes.append(i.RightChild)
        return len(list_nodes)
    
    def StructureCheck (self):
        if self.Root is None:
            return 'Список пуст'
        list_nodes = []
        list_nodes.append(self.Root)
        for i in list_nodes:
            if i.LeftChild is not None:
                list_nodes.append(i.LeftChild)
            if i.RightChild is not None:
                list_nodes.append(i.RightChild)
                
        for i in list_nodes:
            if i.LeftChild is not None:
                if i.LeftChild.NodeKey > i.LeftChild.Parent.NodeKey:
                    print ('Левый узел больше родителя!!!!!!')
                    return
            elif i.RightChild is not None:    
                if i.RightChild.NodeKey < i.RightChild.Parent.NodeKey:
                    print ('Правый узел меньше родителя!!!!!!')
                    return
            if i.Parent is not None:
                    if i.NodeKey < i.Parent.NodeKey:
                        if i != i.Parent.LeftChild:
                            print ('Родительская ссылка поломана!!!!!!!!')
                            return
                    elif i.NodeKey > i.Parent.NodeKey:
                        if i != i.Parent.RightChild:
                            print ('Родительская ссылка поломана!!!!!!!!')
                            return
        return print ('Структурный тест пройден')
    
    def WideAllNodes(self):
        list_nodes = []
        if self.Root is None:
            return list_nodes    
        
        list_nodes.append(self.Root)
        for i in list_nodes:
            if i.LeftChild is not None:
                list_nodes.append(i.LeftChild)
            if i.RightChild is not None:
                list_nodes.append(i.RightChild)

        return list_nodes
    
    def DeepAllNodes(self, order, node = None, list_nodes = [], started = True, left = True):
        if started:
            node = self.Root.LeftChild
            list_nodes.append(node)
        if order == 0:
            if node is not None:
                if left:
                    list_nodes.append(node.LeftChild)
                    left = False
                else:
                    list_nodes.append(node.RightChild)
                    node = node.LeftChild
                    left = True
                self.DeepAllNodes(0, node, list_nodes, False, left)
                
                
            return list_nodes
```
